```python
# 2. Discretize the data.
#    Convert the each (latitude, longitude) pair into a discrete integer, 
#    which we can consider to be a nominal category. You will do this using two functions, 
#    both of which have the following inputs and output:
#        inputs:  'data': nx2 numpy array. The raw latitude and longitude columns from the original data,
#                                         in the original order.
#                 'H':    integer.         The number of bins of height/latitude, across the range of all latitudes
#                 'W':    integer.         The number of bins of width/longitude, across the range of all longitudes
#        
#        outputs: 'out':  list of ints.    The ith element is the index of the bin in the HxW grid, 
#                                         of the ith row in the given 'data', computed H first, e.g.:
#                                         | 0 | 3 | 6 |  9 |
#                                         | 1 | 4 | 7 | 10 |
#                                         | 2 | 5 | 8 | 11 |
#                                         for H=3 and W=4
#
#     First, use the 'equal width' method. See, e.g., Lecture 3, slide 40.
#     
def equalWidth(data: np.ndarray, H: int, W: int):
    # This function uses the 'equal width' method of binning,
    # where the range of each bin is the same across each dimension independently
    out = []
    
    # Compute bin edges for latitude and longitude
    lat_bins = np.linspace(np.min(data[:, 0]), np.max(data[:, 0]), num=H, endpoint=False)
    lon_bins = np.linspace(np.min(data[:, 1]), np.max(data[:, 1]), num=W, endpoint=False)
    
    # Assign data points to bins
    for lat, lon in data:
        lat_index = np.searchsorted(lat_bins, lat)
        lon_index = np.searchsorted(lon_bins, lon)
        bin_index = lat_index * W + lon_index
        out.append(bin_index)
    
    return out
```

```python
# 3. Discretize the data with the 'equal depth' method
#    The same as above, but with the 'equal depth' method

def equalDepth( data:np.ndarray, H:int, W:int ):
    # this function uses the 'equal depth' method of binning, where the number of elements in each bin
    # is approximately same. Given any remainder 'r' of N / (HxW), 'r' bins will have one more element 
    # than the rest.
    out = []
    # TODO: your code goes here
    # Compute the number of elements per bin
    total_bins = H * W
    num_elements = len(data)
    elements_per_bin = num_elements // total_bins
    remainder = num_elements % total_bins
    
    # Sort the data based on latitude
    data_sorted = data[data[:, 0].argsort()]
    
    # Divide the data into bins
    for i in range(total_bins):
        start_index = i * elements_per_bin + min(i, remainder)
        end_index = (i + 1) * elements_per_bin + min(i + 1, remainder)
        lat, lon = data_sorted[start_index:end_index, 0], data_sorted[start_index:end_index, 1]
        lat_index = np.digitize(lat, np.unique(lat)) - 1
        lon_index = np.digitize(lon, np.unique(lon)) - 1
        bin_index = lat_index * W + lon_index
        out.extend(bin_index.tolist())
    
    return out
```

```python
# 2. Code the recursive BUC function 
#        inputs:  'input':     pandas df      Previously sorted data (see above).
#                 'curState':  list of str    The current state, as described above
#                 'dim':       integer.       The starting dimension for this iteration
#        globals: 'numDims':   integer.       The total number of dimensions
#                 'card':      list of ints.  Where card[d] is the cardinality of dimension d
#                 'minSup':    integer.       The minimum support
#                 'dataCount': list of lists. dataCount[d] is a list of integers, each of size card[d]
#                                             dataCount[d][i] is the size of a partition for variable 'i'
#        output:  

# TODO: you can create any helper functions you want here
def compute_measure(cur_state):
    # Compute the measure of the current state
    # Here, for simplicity, let's compute the sum of counts
    measure = 0
    for i, dim_value in enumerate(cur_state):
        measure += dataCount[i][dim_value]
    return measure

def BUC(input_data, curState, dim):
    global numDims, card, minSup, dataCount
    
    if dim == numDims:  # Base case: All dimensions processed
        # Compute measure for the current state
        measure = compute_measure(curState)
        if measure >= minSup:  # Check if the measure meets minimum support
            # Return current state and its measure
            return [{'cell': curState, 'measure': measure}]
        else:
            return []
    
    results = []
    
    # Include current dimension in the state
    for dim_value in range(card[dim]):
        curState[dim] = dim_value
        results += BUC(input_data, curState[:], dim + 1)
    
    # Exclude current dimension in the state
    curState[dim] = -1
    results += BUC(input_data, curState[:], dim + 1)
    
    return results

```
